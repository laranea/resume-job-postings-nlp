```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import os
```

```python
def load_webpages(dir):
    files = os.listdir(dir)
    read_files = []
    print("Extracting", len(files), "html files...")
    for file in files:
        with open(dir + file) as f:
            read_files.append(f.read())
    
    return np.array(read_files)

files = load_webpages("/Users/nicklittlefield/Desktop/Projects/resume-job-postings-nlp/data/html_job_postings/")
```

```python
postings = pd.DataFrame(columns = ["title", "body", "bullets"])
```

```python
def extract_web_components(files):
    titles = []
    bodies = []
    bullets = []
    for file in files:
        soup = bs(file)
        titles.append(soup.find("title").text)
        bodies.append(soup.find("body").text)
        bullets.append([bull.text.strip() for bull in soup.find_all("li")])
    
    return (titles, bodies, bullets)
```

```python
postings["title"], postings["body"], postings["bullets"] = extract_web_components(files)
```

```python
postings['bullets'] = postings['bullets'].apply(tuple, 1)
postings.drop_duplicates(inplace=True)
```
